**1: Write a Python function called recursive_factorial that calculates the factorial of a number using recursion. Ensure it handles edge cases like negative numbers and zero.**

```python
def recursive_factorial(n):
    if n==0:
        return 1
    return n*recursive_factorial(n-1)

print(recursive_factorial(5))
```

**2. Write a Python program to generate the first 10 prime numbers using a while loop and conditions.Optimize the solution to skip even numbers after 2.**

```python
def prime(n):
    if n<2:
        return False
    if n==2:
        return True
    if n%2==0:
        return False
        
    for i in range(3,int(n**0.5),2):
        if n%i==0:
            return False
        return True
        
p=[]
n=2

while len(p)<10:
    if prime(n):
        p.append(n)
    if n==2:
        n=n+1
    else:
        n=n+2
        
print(p)
```

**3. Write a Python program to find the longest substring without repeating characters in a given string. For example, for the input "abcabcbb", the output should be "abc".**

```python
def longest_unique_substring(s):
    char_index = {}  
    start = 0  
    max_length = 0  
    longest_substring = ""  
    
    for end in range(len(s)):
        if s[end] in char_index and char_index[s[end]] >= start:
            start = char_index[s[end]] + 1  
        
        char_index[s[end]] = end  
        
        if end - start + 1 > max_length:
            max_length = end - start + 1
            longest_substring = s[start:end + 1] 
    
    return longest_substring


s = "abcabcbb"
print(longest_unique_substring(s))  
```

[{"name": "Alice", "age": 25}, {"name": "Bob", "age": 30}] by the "age" key.**

**5. Write a lambda function that takes three arguments (x, y, z) and returns the sum of x and y if z is True,otherwise returns the product of x and y.**

```python
sum_or_product = lambda x, y, z: x + y if z else x * y

print(sum_or_product(3, 4, True))  
print(sum_or_product(3, 4, False))
```

**6. Use the map() function to convert a list of temperatures in Celsius [0, 10, 20, 30, 40] to Fahrenheit using the formula F = (C * 9/5) + 32.**

```python
celsius_temperatures = [0, 10, 20, 30, 40]
fahrenheit_temperatures = list(map(lambda c: (c * 9/5) + 32, celsius_temperatures))
print(fahrenheit_temperatures)
```

**7. Use the filter() function to extract all palindromic numbers from a list of integers. For example, from [121, 123, 1331, 456], the output should be [121, 1331].**

```python
numbers = [121, 123, 1331, 456, 989, 1001]
palindromic_numbers = list(filter(lambda num: str(num) == str(num)[::-1], numbers))
print(palindromic_numbers)
```

**8. Use the reduce() function to find the greatest common divisor (GCD) of a list of numbers. For example,for the list [24, 36, 48], the GCD is 12.**

```python
from functools import reduce
import math

numbers = [24, 36, 48]
gcd_result = reduce(math.gcd, numbers)
print(gcd_result)
```

**9. Write a Python function called recursive_power that calculates the power of a number using recursion.For example, recursive_power(2, 3) should return 8. Optimize it to handle negative exponents.**

```python
def recursive_power(base, exponent):
    if exponent == 0:
        return 1  
    if exponent < 0:
        return 1 / recursive_power(base, -exponent)  
    if exponent % 2 == 0:
        half_power = recursive_power(base, exponent // 2)
        return half_power * half_power  
    return base * recursive_power(base, exponent - 1)  


print(recursive_power(2, 3))  
print(recursive_power(2, -3)) 
```
